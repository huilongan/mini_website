```python

import json
import pandas as pd
from glob import glob

def convert(x):
    ob = json.loads(x)
    #print(list(ob.items()))
    for k, v in list(ob.items()):
        #print("start")
        #print(type(v))
        #print(k)
        
        if not isinstance(v, dict):
            ob[k] = ''.join(str(v))
            #print(k)
        elif isinstance(v, dict):
            #print(v)
            for kk, vv in v.items():
                #print(v.items())
                #print(type(kk))
                #print(type(vv))
                ob['%s_%s' % (k, kk)] = vv
                #print('%s_%s' % (k, kk))
            del ob[k]
        #print("end")
    return ob
```

```python
csv_filename = 'yelp_academic_dataset_user.csv'
print ('Converting %s to %s' % ("yelp_academic_dataset_user.json", csv_filename))
df = pd.DataFrame([convert(line) for line in open('yelp_academic_dataset_user.json')])
#df.to_csv(csv_filename, encoding='utf-8', index=False)
```

```python
df.to_csv(csv_filename, encoding='utf-8', index=False)
```
